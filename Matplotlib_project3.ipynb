```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('covid data.csv')
print(df.tail(10))
```

```python
country_data = df[df['location'] == 'India']

plt.plot(country_data['date'], country_data['infected'], color='blue')
plt.title('Total Infected Cases Over Time in India')
plt.xlabel('Date')
plt.ylabel('Infected Cases')


plt.show()
```

```python
continent_deaths = df.groupby('continent')['dead'].sum().reset_index()

plt.bar(continent_deaths['continent'], continent_deaths['dead'], color='r')
plt.title('Total Deaths by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Deaths')
plt.show()
```

```python
continent_data = df.groupby('continent')[['infected', 'vaccinated']].sum().reset_index()

plt.bar(continent_data['continent'], continent_data['infected'], color='orange', label='Infected')
plt.bar(continent_data['continent'], continent_data['vaccinated'], bottom=continent_data['infected'], color='green', label='Vaccinated')
plt.title('Infected vs Vaccinated by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Numbers')
plt.legend()
plt.show()
```

```python
date_data = df[df['date'] == '4/26/2021']

plt.hist(date_data['dead'], bins=20,color='red', edgecolor='black')
plt.title('Histogram of Deaths Across Countries on 2022-01-01')
plt.xlabel('Deaths')
plt.ylabel('Number of Countries')
plt.show()
```

```python
date_data = df[df['date'] == '4/26/2021']

plt.scatter(date_data['infected'], date_data['dead'], color='purple')
plt.title('Infected vs Deaths on 2021-12-01')
plt.xlabel('Infected Cases')
plt.ylabel('Deaths')
plt.show()
```

```python
continent_data = df[df['continent'] == 'Asia']

cumulative_vaccination = continent_data.groupby('date')['vaccinated'].sum()

plt.fill_between(cumulative_vaccination.index, cumulative_vaccination.values, color='green', alpha=0.6)
plt.title('Cumulative Vaccinated Over Time in Asia')
plt.xlabel('Date')
plt.ylabel('Cumulative Vaccinated')
plt.show()
```
